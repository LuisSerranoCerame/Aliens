# Alien Films

```python
dffilms = pd.read_csv('F_films.csv')
```

```python
dffilms.value_counts()
```

```python
dffilms['Year'] = pd.to_datetime(dffilms['Year'], format="%Y")
```

```python
dffilms.dtypes
```

```python
def get_year_film (mi_fecha):
    try:
        cond = dffilms['Year'] < mi_fecha
        df_past = dffilms[cond]
        dfn_past = df_past.sort_values(['Year'])
        return dfn_past["Year"].iloc[-1]
    except:
        return None
```

```python
df["YearFilm"] = df["datetime"].apply(get_year_film)
```

```python
def get_film (mi_fecha):
    try:
        cond = dffilms['Year'] < mi_fecha
        df_past = dffilms[cond]
        dfn_past = df_past.sort_values(['Year'])
        return dfn_past["Title"].iloc[-1]
    except:
        return None
```

```python
df["Film"] = df["datetime"].apply(get_film)
```
