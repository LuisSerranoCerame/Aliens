```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
from selenium.webdriver.support.ui import Select
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
opciones=Options()
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.headless=False    # si True, no aperece la ventana (headless=no visible)
opciones.add_argument('--start-maximized')         # comienza maximizado
#opciones.add_argument('user-data-dir=selenium')    # mantiene las cookies
#opciones.add_extension('driver_folder/adblock.crx')       # adblocker
opciones.add_argument('--incognito')              # incognito
```

```python
year = []
milliondollars = []
perfederalbudget = []

driver = "./chromedriver.exe"
driver = webdriver.Chrome(driver, options = opciones)
driver.get("https://en.wikipedia.org/wiki/Budget_of_NASA")

for i in range(3,47):
    year.append(driver.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table/tbody/tr/td[1]/table/tbody/tr[{i}]/td[1]').text)
    milliondollars.append(driver.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table/tbody/tr/td[1]/table/tbody/tr[{i}]/td[2]').text)
    perfederalbudget.append(driver.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table/tbody/tr/td[1]/table/tbody/tr[{i}]/td[3]').text)
```

```python
firstpart = pd.DataFrame(list(zip(year,milliondollars,perfederalbudget)), columns = ['Year','Nominal_Dollars(Millions)','%Federal_Budget'])
```

```python
firstpart
```

```python
year2 = []
milliondollars2 = []
perfederalbudget2 = []

driver = "./chromedriver.exe"
driver = webdriver.Chrome(driver, options = opciones)
driver.get("https://en.wikipedia.org/wiki/Budget_of_NASA")

for i in range(3,33):
    year2.append(driver.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table/tbody/tr/td[2]/table/tbody/tr[{i}]/td[1]').text)
    milliondollars2.append(driver.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table/tbody/tr/td[2]/table/tbody/tr[{i}]/td[2]').text)
    perfederalbudget2.append(driver.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table/tbody/tr/td[2]/table/tbody/tr[{i}]/td[3]').text)
```

```python
secondpart = pd.DataFrame(list(zip(year2,milliondollars2,perfederalbudget2)), columns = ['Year','Nominal_Dollars(Millions)','%Federal_Budget'])
```

```python
secondpart
```

```python
year3 = []
milliondollars3 = []
perfederalbudget3 = []

driver = "./chromedriver.exe"
driver = webdriver.Chrome(driver, options = opciones)
driver.get("https://en.wikipedia.org/wiki/Budget_of_NASA")

for i in range(3,33):
    year3.append(driver.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table/tbody/tr/td[3]/table/tbody/tr[{i}]/td[1]').text)
    milliondollars3.append(driver.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table/tbody/tr/td[3]/table/tbody/tr[{i}]/td[2]').text)
    perfederalbudget3.append(driver.find_element_by_xpath(f'//*[@id="mw-content-text"]/div[1]/table/tbody/tr/td[3]/table/tbody/tr[{i}]/td[3]').text)
```

```python
newm= []
separador = "["
for i in milliondollars3:
    newm.append(milliondollars3[i].split(separador)
```

```python
newm2 = []
for i in range(len(milliondollars3)):
    newm2.append(milliondollars3[i][0:6])
```

```python
newm2
```

```python
thirdpart = pd.DataFrame(list(zip(year3,newm2,perfederalbudget3)), columns = ['Year','Nominal_Dollars(Millions)','%Federal_Budget'])
```

```python
thirdpart
```

```python
nasa_yearly_budget = pd.concat([firstpart,secondpart, thirdpart], axis=0)
```

```python
nasa_yearly_budget.reset_index()
```
